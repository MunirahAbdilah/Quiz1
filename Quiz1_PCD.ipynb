```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('/content/lena.jpg')

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_img, cmap='gray')
plt.show()
```

# Cari dimensi gambar


```python
height, width = gray_img.shape[:2]
print("Width:", width)
print("Height:", height)
```

# Nilai Max & Min

```python
max_value = np.max(gray_img)
min_value = np.min(gray_img)
print("Max Value:", max_value)
print("Min Value:", min_value)
```

# Normalisasi

```python
normalized_img = (gray_img - min_value) / (max_value - min_value)

plt.imshow(normalized_img, cmap='gray')
plt.title('Normalized Image')
plt.show()
```

# Flip horizontal & vertikal

```python
flipped_horizontal = cv2.flip(gray_img, 1)

flipped_vertical = cv2.flip(gray_img, 0)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(flipped_horizontal, cmap='gray')
plt.title('Flipped Horizontal')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(flipped_vertical, cmap='gray')
plt.title('Flipped Vertical')
plt.axis('off')

plt.show()
```
